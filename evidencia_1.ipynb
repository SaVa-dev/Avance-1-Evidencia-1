```python
ruta_archivo = 'Tarifas por puntos 2016-2017.csv'
archivo_csv = pd.read_csv(ruta_archivo)

archivo_csv
```

```python
zona = archivo_csv['zona_inyeccion'].tolist()
capacidad = archivo_csv['capacidad_base_firme'].tolist()

zonas_capacidad_firme = {'Norte' : 0,
         'Sur' : 0,
         'Occidente' : 0, 
         'Golfo' : 0 
         }

for i in range(0, len(capacidad)):
        zonas_capacidad_firme[zona[i]] += capacidad[i]
    
```

```python
fig = plt.figure(figsize = (5, 5))

region = list(zonas_capacidad_firme.keys())
resultados = list(zonas_capacidad_firme.values())
plt.bar(region, resultados, width = 0.5)
plt.show()
```
